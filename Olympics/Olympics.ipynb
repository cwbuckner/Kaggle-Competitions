Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/12th Grade/Machine Learning/Data/athlete_events.csv')
```

```python
plt.figure(figsize=(16, 8))  # Increase the figure size for better readability
sns.set(style="whitegrid")
sns.countplot(x='Sport', hue='Medal', data=df, palette='Set1')
plt.title('Medals won by Countries in Different Sports')
plt.xlabel('Sport')
plt.ylabel('Count')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.legend(title='Medal', loc='upper right', labels=['Gold', 'Silver', 'Bronze'])
plt.tight_layout()  # Add this to prevent labels from getting cut off
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the count of medals by Gender
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sex', hue='Medal', palette='Set1')
plt.title('Count of Medals by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Medal', loc='upper right', labels=['Gold', 'Silver', 'Bronze'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Height', y='Weight', hue='Medal', palette='Set1')
plt.title('Scatter Plot of Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend(title='Medal', loc='upper right', labels=['Gold', 'Silver', 'Bronze'])
plt.show()
```

```python
numerical_vars = ['Age', 'Height', 'Weight']
sns.set_theme(style="ticks")
sns.pairplot(df[numerical_vars + ['Medal']], hue='Medal', palette='Set1')
plt.show()
```

Altair

```python
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/12th Grade/Machine Learning/Data/athlete_events.csv')
```

```python
import altair as alt

# Load your dataset (replace 'your_data.csv' with the actual path to your data)

# Sample a subset of your data (adjust 'sample_size' as needed)
sample_size = 500  # Choose an appropriate sample size
sampled_data = df.sample(sample_size, random_state=42)

# Visualization code (replace 'sampled_data' with the sampled dataset)
scatter = alt.Chart(sampled_data).mark_circle(size=60).encode(
    x='Age:Q',
    y='Height:Q',
    color='Medal:N',
    tooltip=['Name', 'Age', 'Height', 'Medal']
).properties(
    width=500,
    height=300
)

scatter
```

```python
# Sample a subset of your data (adjust 'sample_size' as needed)
sample_size_sport = 500  # Choose an appropriate sample size
sampled_df_sport = df.sample(sample_size_sport, random_state=42)

bar_chart_sport = alt.Chart(sampled_df_sport).mark_bar().encode(
    x='count():Q',
    y=alt.Y('Sport:N', sort='-x'),
    color='Medal:N',
    tooltip=['Sport', 'count()']
).properties(
    width=500,
    height=400
)

bar_chart_sport
```

```python
# Sample a subset of your data (adjust 'sample_size' as needed)
sample_size_boxplot = 500  # Choose an appropriate sample size
sampled_df_boxplot = df.sample(sample_size_boxplot, random_state=42)

boxplot_weight_sport = alt.Chart(sampled_df_boxplot).mark_boxplot().encode(
    x='Sport:N',
    y='Weight:Q',
    color='Sport:N'
).properties(
    width=600,
    height=400,
    title='Weight Distribution by Sport (Box Plot)'
)

boxplot_weight_sport
```

```python
# Specify the sample size
sample_size = 100  # Adjust the sample size as needed

# Sample a subset of the data
sampled_df = df.sample(sample_size, random_state=42)

# Create a categorical scatterplot using Altair
scatter = alt.Chart(sampled_df).mark_circle(size=60).encode(
    x='body_mass_g:Q',
    y='sex:N',
    color='team:N'
).properties(
    width=400,
    height=200
)

scatter
```

```python
import altair as alt
import pandas as pd

# Select a specific sport (you can change this to any sport in your dataset)
selected_sport = 'Swimming'

# Filter the data for the selected sport and medalists
filtered_df = df[(df['Sport'] == selected_sport) & (df['Medal'].isin(['Gold', 'Silver', 'Bronze']))]

# Create a histogram using Altair
histogram_age_medalists = alt.Chart(filtered_df).mark_bar().encode(
    x=alt.X('Age:Q', bin=alt.Bin(maxbins=20), axis=alt.Axis(title='Age')),
    y=alt.Y('count():Q', axis=alt.Axis(title='Count')),
    color=alt.Color('Medal:N', scale=alt.Scale(domain=['Gold', 'Silver', 'Bronze'], range=['gold', 'silver', 'brown'])),
    tooltip=['Medal:N', 'Age:Q', 'count():Q']
).properties(
    width=600,
    height=400,
    title=f'Age Distribution of Medalists in {selected_sport}'
).configure_title(
    fontSize=14,
    anchor='middle'
)

histogram_age_medalists
```

```python
#ample data (you can replace this with your own dataset)
data = pd.DataFrame({
    'Age': np.random.randint(20, 40, 200),  # Random ages between 20 and 40
    'Height': np.random.uniform(150, 190, 200),  # Random heights between 150 and 190
    'Medal': np.random.choice(['Gold', 'Silver', 'Bronze'], size=200)  # Random medals
})

# Create an interactive scatter plot using Altair with 'Height' on the y-axis
scatter_plot = alt.Chart(data).mark_circle().encode(
    x='Age:Q',
    y='Height:Q',  # Change y-axis label to 'Height'
    size=alt.Size('Height:Q', scale=alt.Scale(domain=[150, 190])),
    color='Medal:N',
    tooltip=['Age:Q', 'Height:Q', 'Medal:N']
).interactive()

scatter_plot
```
